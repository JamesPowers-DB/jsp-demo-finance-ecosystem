```python
from pyspark.sql.types import (
    StructType,
    StructField,
    StringType,
    LongType,
    DoubleType,
    BooleanType
    )

    

invoice_schema = StructType([
    StructField("invoice_id", LongType(), nullable=False),
    StructField("purchase_order_id", LongType(), nullable=True),
    StructField("invoice_number", StringType(), nullable=False),
    StructField("invoice_date", LongType(), nullable=True),
    StructField("invoice_status", StringType(), nullable=True),
    StructField("supplier_id", LongType(), nullable=True),
    StructField("invoice_amount", DoubleType(), nullable=True),
    StructField("tax_amount", DoubleType(), nullable=True),
    StructField("discount_amount", DoubleType(), nullable=True),
    StructField("total_invoice_amount", DoubleType(), nullable=True),
    StructField("amount_paid", DoubleType(), nullable=True),
    StructField("payment_due_date", LongType(), nullable=True),
    StructField("invoice_category", StringType(), nullable=True),
    StructField("payment_date", LongType(), nullable=True),
    StructField("payment_method", StringType(), nullable=True),
    StructField("payment_reference", StringType(), nullable=True),
    StructField("receipt_matched", BooleanType(), nullable=True),
    StructField("three_way_match_status", StringType(), nullable=True),
    StructField("goods_received_date", LongType(), nullable=True),
    StructField("coa_id", LongType(), nullable=False)
    ])

path = '/Volumes/jsp_demo/fin/data_gen_outputs/spend_invoices/'

inv_df = spark.read\
    .option("recursiveFileLookup", "true")\
    .csv(
        path=path,
        header=True,
        schema=invoice_schema,
    )

inv_df.display(100)
```

```python
%sql 
SELECT 
  YEAR(FROM_UNIXTIME(invoice_date)), 
  purchase_order_id IS NOT NULL, 
  COUNT(*), 
  SUM(total_invoice_amount) AS inv_amount
FROM jsp_demo.spend_bronze.raw_spend_invoices GROUP BY 1, 2
ORDER BY 2 ,1 
```

```python
%sql 
SELECT 
  YEAR(FROM_UNIXTIME(transaction_date)), 
  -- po_id IS NOT NULL, 
  COUNT(*), 
  SUM(amount) AS inv_amount
FROM jsp_demo.rev_bronze.raw_revenue_transactions GROUP BY 1
ORDER BY 1 
```

```python
%sql 
SELECT 
  YEAR(FROM_UNIXTIME(invoice_date)) AS yr, 
  COUNT(*), 
  SUM(billed_amount) AS inv_amount
FROM jsp_demo.rev_bronze.raw_revenue_billings GROUP BY 1
ORDER BY 1 
```
