```python
%sql
SELECT 
    COUNT(third_party_id), 
    COUNT(DISTINCT third_party_id), 
    COUNT(supplier_id), 
    COUNT(DISTINCT supplier_id), 
    COUNT(customer_id), 
    COUNT(DISTINCT customer_id)
FROM jsp_demo.tp_gold.dim_third_party

```

```python
%sql
SELECT
  third_party_id
  ,ARRAY_AGG(customer_name)
FROM jsp_demo.tp_bronze.raw_customers
WHERE third_party_id IN (SELECT third_party_id FROM jsp_demo.tp_bronze.raw_suppliers)
GROUP BY 1
HAVING COUNT(DISTINCT customer_id) > 1
```

```python
%sql
SELECT
  third_party_id
  ,ARRAY_AGG(supplier_name)
FROM jsp_demo.tp_bronze.raw_suppliers
GROUP BY 1
HAVING COUNT(DISTINCT supplier_id) > 1
```
