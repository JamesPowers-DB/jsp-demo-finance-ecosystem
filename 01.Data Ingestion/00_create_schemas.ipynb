```python
%sql
USE CATALOG fin_demo;

/* -- REVENUE SCHEMAS  -- */
CREATE SCHEMA IF NOT EXISTS fin_demo.fin;
CREATE SCHEMA IF NOT EXISTS fin_demo.acct;
CREATE SCHEMA IF NOT EXISTS fin_demo.hr;
CREATE SCHEMA IF NOT EXISTS fin_demo.legal;
CREATE SCHEMA IF NOT EXISTS fin_demo.party;
CREATE SCHEMA IF NOT EXISTS fin_demo.rev;
CREATE SCHEMA IF NOT EXISTS fin_demo.spend;
```

```python
%sql
DESCRIBE fin_demo.spend.stg_spend_transactions;
```

```python
from pyspark.sql.functions import lit

spend_df = spark.sql("DESCRIBE fin_demo.spend.stg_spend_transactions")
spend_df = spend_df.withColumn('table_source', lit('stg_spend_transactions'))

rev_df = spark.sql("DESCRIBE fin_demo.rev.stg_revenue_transactions;")
rev_df = rev_df.withColumn('table_source', lit('stg_revenue_transactions'))

contract_df = spark.sql("DESCRIBE fin_demo.legal.dim_all_contracts;")
contract_df = contract_df.withColumn('table_source', lit('dim_all_contracts'))

df = spend_df.union(rev_df)
df = df.union(contract_df)
df = df.drop('comment')

df.display()
```
