```python
%sql

SELECT * FROM fin_demo.dbx.raw_customer_matrix
WHERE anon_name IN (SELECT anon_name FROM fin_demo.dbx.raw_customer_matrix GROUP BY 1 HAVING COUNT(*) > 1)
ORDER BY anon_name, account_id
```
